```python

def cos_figsize(t):
    tmp_f = 1.
    tmp_val = np.cos(tmp_f * t)
    return tmp_val

def cos2_sig035_figsize(t):
    tmp_f0 = 2.
    tmp_f1 = .35
    tmp_val = np.cos(tmp_f0 * t + np.sin(tmp_f1 * t))
    return tmp_val

def cos03_cos065_figsize(t):
    tmp_f0 = .3
    tmp_f1 = 0.65
    tmp_val = np.cos(tmp_f0 * t + np.cos(tmp_f1 * t))
    return tmp_val

def cos03_tan065_figsize(t):
    tmp_f0 = 0.3
    tmp_f1 = 0.65
    tmp_val = np.cos(tmp_f0 * t + np.tan(tmp_f1 * t))
    return tmp_val

def cos03_tan05_figsize(t):
    tmp_f0 = 0.3
    tmp_f1 = 0.5
    tmp_val = np.cos(tmp_f0 * t + np.tan(tmp_f1 * t))
    return tmp_val

def cos05_tan05_figsize(t):
    tmp_f0 = 0.5
    tmp_f1 = 0.5
    tmp_val = np.cos(tmp_f0 * t + np.tan(tmp_f1 * t))
    return tmp_val

def cos_tan_sin_figsize(t):
    tmp_val = np.cos(t + np.tan(t) + np.sin(t))
    return tmp_val
```
