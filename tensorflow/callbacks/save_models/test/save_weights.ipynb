# Import layers

```python
import numpy as np
from tensorflow.keras import layers, Input,  Model
import sys
from pathlib import Path
```

```python

p = Path()
LOCAL_PATH   = str(p.cwd().parent.parent)
MODELS_PATH  = str(p.cwd().parent.joinpath("test", "models"))
DATASET_PATH = str(p.cwd().parent.joinpath("test", "dataset"))
print("local path: {}".format(LOCAL_PATH))
```

```python

# adding local_folder to the system path
sys.path.append(LOCAL_PATH)

def remove_modules(modules_name, *arg):
    if (modules_name in sys.modules):
        del sys.modules[modules_name]
        for key in arg:
            tmp_modules = "{}.{}".format(modules_name, key)
            if (tmp_modules in sys.modules):
                del sys.modules[tmp_modules]
```

```python

remove_modules("save_models.code", "save_weights")

from save_models.code.save_weights import *
```

# Make model

```python

inputs  = Input(shape=(28, 28, 3))
outputs = layers.Dense(1, 
                        activation=None,
                        name="dense",
                    )(inputs)
model1 = Model(inputs, outputs, name="model_1")

inputs  = Input(shape=(28, 28, 3))
outputs = layers.Dense(1, 
                        activation=None,
                        name="dense",
                    )(inputs)
model2 = Model(inputs, outputs, name="model_2")

models = [model1, model2]
```
